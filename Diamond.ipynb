Load dataset into a pandas DataFrame

```python
import pandas as pd

%matplotlib inline

url = "https://raw.githubusercontent.com/amankharwal/Website-data/refs/heads/master/diamonds.csv"
diamonds = pd.read_csv(url, index_col=0)
```

Display the First seven Rows

```python
print(diamonds.head(7))
```

Display the Last seven Rows

```python
print(diamonds.tail(7))
```

Use the DataFrame Method *describe*

```python
diamonds.describe()
```

Use Series Method *describe*

```python
diamonds[['cut', 'color', 'clarity']].describe()
```

Use Series Method ***unique***

```python
print('Cut Categories', diamonds['cut'].unique(), end="\n\n")

print('Color Categories', diamonds['color'].unique(), end="\n\n")

print('Clarity Categories', diamonds['clarity'].unique(), end="\n\n")
```

Use DataFrame *hist* Method

```python
diamonds.hist(figsize=(15, 8))
```
